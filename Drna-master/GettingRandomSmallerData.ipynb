# Picking random data


## Training data (picking 1000 in each category)

```python
pick_in_each_category = 1000
data_root = ""
data_list = "./data/List/train_images_root.txt"
f = open(data_list, 'r')
data_list = f.readlines()
f.close()
data_dict = {"Food" : [], 
             "Clothes": [],
             "Institution" : [],
             "Accessories" : [],
             "Transportation" : [],
             "Electronic" : [],
             "Necessities" : [],
             "Cosmetic" : [],
             "Leisure" : [],
             "Medical" : []
             }
```

```python
for elem in data_list:
    cat = elem.split("/")[0]
    data_dict[cat].append(elem)
```

```python
import numpy as np
np.random.seed(12)
small_data_list = []
for cat in data_dict.keys():
    idx = np.random.randint(0,len(data_dict[cat]), pick_in_each_category)
    chosen = np.array(data_dict[cat])[idx]
    for elem in chosen:
        small_data_list.append(elem)
```

```python
with open("./data/List/train_images_root_small.txt", "w") as f:
  f.writelines(small_data_list)
```
